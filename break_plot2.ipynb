```python
import gzip
import json
import pandas as pd
import datetime
import matplotlib.dates as mdates
import matplotlib.pyplot as plt
import plotly.express as px
import plotly.io as pio
import plotly.graph_objects as go
from IPython.display import HTML # 导入HTML
```

```python
date = datetime.date(2021,9,7)
data = []
file = f'C:/Users/13365/Downloads/{date.strftime("%Y-%m-%d")}.json.gz'    
with gzip.open(file, 'rb') as f:
    for line in f:
        data.append(json.loads(line))
# 将数据转换为DataFrame
df_1 = pd.DataFrame(data)
# 将时间戳转换为日期时间格式
df_1['t'] = pd.to_datetime(df_1['t'])
# 将时间戳列设置为索引
df_1.set_index('t', inplace=True)
```

```python
time2 = df_1.between_time('15:06:00+00:00','15:11:05+00:00')
# 绘制折线图
plt.figure(figsize=(10, 6))
plt.plot(time2.index, time2['r'],linewidth=0.3,color = 'black')
plt.title('Price Over Time')
plt.xlabel('Time')
plt.ylabel('Price')
plt.grid(True)
plt.show()
```
